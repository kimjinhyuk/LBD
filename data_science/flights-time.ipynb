# 파이썬 시계열 데이터 시각화

## 가설을 세우고 검정하기 : 여름철에는 항공이용이 많을 것이다.

### 분석을 통해 얻는 점
  * 추세를 파악할 수 있음
    * 수요가 늘어나고 있다면 늘어나는 수요에 대비
    * 수요가 줄어들고 있다면 서비스 개선을 하거나 새로운 사업계획 수립
  * 계절에 따른 수요계획을 세울 수 있음
    * 재고량, 예산, 운영인력, 시스템을 구성하는 서버의 트래픽 양 등

```python
#라이브러리 로드
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
# 경고메시지 무시
import warnings
warnings.filterwarnings("ignore")
```

```python
#데이터셋 로드하기
df = sns.load_dataset("flights")
df.shape
```

```python
df.info()
```

```python
df.head()
```

## 데이터로 다양한 시각화 해보기
  * bar, line, point, box, violin, swarm plot

```python
#bar plot
plt.figure(figsize=(15,4))
sns.barplot(data=df, x="year", y="passengers" ,ci="sd") #ci 신뢰구간
```

```python
#line plot
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x="year", y="passengers")
```

```python
#point plot
plt.figure(figsize=(15,4))
sns.pointplot(data=df, x="year", y="passengers", ci='sd') #표준편차로는 시간이 갈 수록 월 마다 차이가 많이 난다
```

```python
# barplot 에 hue를 통해 월별 색상 다르게 그리기

plt.figure(figsize=(15,4))
sns.barplot(data=df, x="year", y="passengers", hue='month') 
```

```python
# lineplot 에 hue를 통해 월별 색상 다르게 그리기
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x="year", y="passengers", hue='month') 
plt.legend(bbox_to_anchor=(1.02,1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x="month", y="passengers", hue='year') 
plt.legend(bbox_to_anchor=(1.02,1), loc=2, borderaxespad=0.)
```

```python
#boxplot
plt.figure(figsize=(15,4))
sns.boxplot(data=df, x='year', y='passengers')
```

```python
#boxenplot
plt.figure(figsize=(15,4))
sns.boxenplot(data=df, x="year", y="passengers")
```

```python
# violinplot

plt.figure(figsize=(15,4))
sns.violinplot(data=df, x="year", y="passengers")
#하얀색 점 중앙값
```

```python
#swarmplot

plt.figure(figsize=(15,4))
sns.swarmplot(data=df, x="year", y="passengers")
```

```python
#boxplot
plt.figure(figsize=(15,4))
sns.boxplot(data=df, x='month', y='passengers')
```

```python
#violinplot
plt.figure(figsize=(15,4))
sns.violinplot(data=df, x='month', y='passengers')
```

```python
#regplot
plt.figure(figsize=(15,4))
sns.regplot(data=df, x='year', y='passengers',truncate=False)
```

```python
#swarmplot

plt.figure(figsize=(15,4))
sns.swarmplot(data=df, x="year", y="passengers")
```

### pivot으로 데이터 형태 바꾸기

```python
df_h = df.pivot(index="month", columns="year", values="passengers")
df_h
```

```python
# heatmap
plt.figure(figsize=(15,5))
sns.heatmap(df_h, annot=True, fmt=".0f", cmap="Blues")
```

```python
#distplot

sns.distplot(df['passengers'])
```

```python
# pairplot
sns.pairplot(data=df, hue="month")
```

```python
#relplot
sns.relplot(data=df, x="month", y="passengers",kind='line', hue="year",col="year", col_wrap=4)
```
